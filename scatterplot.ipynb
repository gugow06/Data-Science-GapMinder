## Child mortality (0-5 year-olds dying per 1000 born)
Descição: Taxa de mortalidade de crianças com menos de 5 anos per 1000 nascidos.

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("children.csv")
df
```

```python
x = ['2026', '2027', '2028', '2029', '2030']

paises = [(df[df['country'] == 'Brazil']), (df[df['country'] == 'India']), (df[df['country'] == 'Russia']), 
          (df[df['country'] == 'South Africa']), (df[df['country'] == 'China'])]

fertilidade_brazil = [paises[0]['2026'].values[0], paises[0]['2027'].values[0], paises[0]['2028'].values[0], 
                      paises[0]['2029'].values[0], paises[0]['2030'].values[0]]

fertilidade_india = [paises[1]['2026'].values[0], paises[1]['2027'].values[0], paises[1]['2028'].values[0], 
                      paises[1]['2029'].values[0], paises[1]['2030'].values[0]]

fertilidade_russia = [paises[2]['2026'].values[0], paises[2]['2027'].values[0], paises[2]['2028'].values[0], 
                      paises[2]['2029'].values[0], paises[2]['2030'].values[0]]

fertilidade_southafrica = [paises[3]['2026'].values[0], paises[3]['2027'].values[0], paises[3]['2028'].values[0], 
                      paises[3]['2029'].values[0], paises[3]['2030'].values[0]]

fertilidade_china = [paises[4]['2026'].values[0], paises[4]['2027'].values[0], paises[4]['2028'].values[0], 
                      paises[4]['2029'].values[0], paises[4]['2030'].values[0]]

plt.scatter(x, fertilidade_brazil, s=100, c='yellow', alpha=0.5, label='Brazil')
plt.scatter(x, fertilidade_india, s=100, c='purple', alpha=0.5, label='India')
plt.scatter(x, fertilidade_russia, s=100, c='blue', alpha=0.5, label='Russia')
plt.scatter(x, fertilidade_southafrica, s=100, c='black', alpha=0.5, label='South Africa')
plt.scatter(x, fertilidade_china, s=100, c='red', alpha=0.5, label='China')

plt.title("Taxa de Fertilidade - BRICS - 2026 até 2030 ")
plt.xlabel("Ano")
plt.ylabel("Taxa de Fertilidade")

plt.legend()

plt.show()
```
